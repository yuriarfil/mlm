<h2>Weather prediction</h2>

```python
% matplotlib inline
import numpy as np
import pandas as pd
import sklearn
```

```python
# load data
df = pd.read_csv('data/nb.csv')
```

```python
# data before encode
df
```

```python
# encode labels with value between 0 and n_classes-1
from sklearn.preprocessing import LabelEncoder

number = LabelEncoder()
df['Outlook'] = number.fit_transform(df['Outlook'])
df['Temperature'] = number.fit_transform(df['Temperature'])
df['Humidity'] = number.fit_transform(df['Humidity'])
df['Windy'] = number.fit_transform(df['Windy'])
df['Play'] = number.fit_transform(df['Play'])
```

```python
# define the features and the target variables
x = ["Outlook", "Temperature", "Humidity", "Windy"]
y = "Play"
```

```python
# create a train, test split. build the model using the train dataset it will validate the model on the test dataset.
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

features_train, features_test, target_train, target_test = train_test_split(df[x],
df[y],
test_size = 0.33,
random_state = 42)
```

```python
# naive bayes(gaussian)
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(features_train, target_train)
```

```python
# prediction and accuracy
pred = clf.predict(features_test)
accuracy = accuracy_score(target_test, pred)
```

```python
# data after encode
df
```

```python
# predict new data
# (sunny, hot, normal, false)
print(clf.predict([[2, 1, 1, 0]]))

clf_pf = GaussianNB()
clf_pf.partial_fit(features_train, target_train, np.unique(target_train))

print(clf_pf.predict([[2, 1, 1, 0]]))
```
